Enumerar los archivos CSV de la carpeta de Drive especificada

```python
import os

drive_path = '/content/drive/MyDrive/MaestríaArqSw/DiseñoOptimizacionBD'

# Check archivos
if os.path.exists(drive_path):
    print(f"Listado de archivos: {drive_path}")
    csv_files = [f for f in os.listdir(drive_path) if f.endswith('.csv')]
    if csv_files:
        print("Archivos CSV encontrados:")
        for f in csv_files:
            print(f"- {f}")
    else:
        print("No se encntraron archivos CSV en este directorio.")
else:
    print(f"Error: Directorio no encontrado {drive_path}")
    csv_files = []
```

Carga y realiza un análisis EDA básico para cada archivo CSV
Se cargan en un DataFrame de pandas y muestra las filas (df.head()), su información (df.info()) y sus estadísticas descriptivas (df.describe()).


```python
import pandas as pd

for csv_file_name in csv_files:
    file_path = os.path.join(drive_path, csv_file_name)
    print(f"\n--- Procesando archivo: {csv_file_name} ---")
    try:
        df = pd.read_csv(file_path)
        print(f"DataFrame '{csv_file_name}' carga exitosa.")

        print("\nCabeceras del DataFrame:")
        display(df.head())

        print("\nInformacion sobre el DataFrame (data types, non-null counts):")
        df.info()

        print("\nDescripción:")
        display(df.describe(include='all'))

    except Exception as e:
        print(f"Error cargando o procesando {csv_file_name}: {e}")
```

Identificación de relaciones entre tablas


```python
import pandas as pd
import os

dfs = {}

print("--- Listado de columnas de cada DataFrame ---")
for csv_file_name in csv_files:
    df_name = os.path.splitext(csv_file_name)[0]
    file_path = os.path.join(drive_path, csv_file_name)
    try:
        df = pd.read_csv(file_path)
        dfs[df_name] = df
        print(f"\nColumnas en '{df_name}':")
        for col in df.columns:
            print(f"- {col}")
    except Exception as e:
        print(f"Error cargando {csv_file_name}: {e}")
```

Buscar columnas comunes en varios DataFrames para identificar posibles claves de unión.

```python
from collections import defaultdict

column_presence = defaultdict(list)

for df_name, df in dfs.items():
    for col in df.columns:
        column_presence[col].append(df_name)

print("--- Se identificaron posibles relaciones basadas en columnas comunes. ---")
for col, df_list in column_presence.items():
    if len(df_list) > 1:
        print(f"Relación potencial a través de la columna '{col}' entre: {', '.join(df_list)}")
```

Análisis de la calidad de los datos: valores nulos

```python
import pandas as pd
import os

print("--- Realización de análisis de calidad de datos ---")

# Re-defining drive_path and populating csv_files (from cell fa0b97b4)
drive_path = '/content/drive/MyDrive/MaestríaArqSw/DiseñoOptimizacionBD'

csv_files = []
if os.path.exists(drive_path):
    csv_files = [f for f in os.listdir(drive_path) if f.endswith('.csv')]
    if not csv_files:
        print("Advertencia: No se encontraron archivos CSV en el directorio especificado.")
else:
    print(f"Error: Directorio no encontrado {drive_path}. No se pudieron cargar los archivos CSV.")

# Re-creating dfs dictionary (from cell 21d57dbf)
dfs = {}
if csv_files:
    for csv_file_name in csv_files:
        # df_name for dfs keys should be without the .csv extension
        df_key = os.path.splitext(csv_file_name)[0]
        file_path = os.path.join(drive_path, csv_file_name)
        try:
            df = pd.read_csv(file_path)
            dfs[df_key] = df
        except Exception as e:
            print(f"Error cargando {csv_file_name}: {e}")
else:
    print("No hay archivos CSV para cargar en el diccionario 'dfs'.")

# List of file names to process (original names from the prompt)
df_names_files = [
    'product_category_name_translation.csv',
    'olist_orders_dataset.csv',
    'olist_geolocation_dataset.csv',
    'olist_order_reviews_dataset',
    'olist_products_dataset',
    'olist_sellers_dataset',
    'olist_customers_dataset',
    'olist_order_payments_dataset',
    'olist_order_items_dataset'
]

# Iterate through the specified list of DataFrame names
for original_file_name in df_names_files:
    # Get the key for the dfs dictionary by stripping the .csv extension
    df_key = os.path.splitext(original_file_name)[0]

    if df_key in dfs: # Check if the DataFrame is loaded in dfs
        df = dfs[df_key]
        print(f"\n--- Analyzing DataFrame: '{original_file_name}' ---")

        # Missing Values Analysis
        missing_values = df.isnull().sum()
        missing_percentage = (df.isnull().sum() / len(df)) * 100
        missing_info = pd.DataFrame({
            'Cuenta faltante': missing_values,
            'Porcentaje faltante': missing_percentage
        })
        missing_info = missing_info[missing_info['Cuenta faltante'] > 0]

        if not missing_info.empty:
            print("\nValores perdidos:")
            display(missing_info.sort_values(by='Porcentaje faltante', ascending=False))
        else:
            print("\nNo se encontraron valores faltantes.")

        # Data Types
        print("\nTipos de Datos:") # Fixed: changed \T to \n
        print(df.dtypes)

        # Unique Values Count
        print("\nValores únicos por columna:")
        for col in df.columns:
            num_unique = df[col].nunique()
            print(f"- {col}: {num_unique} Valores unicos")
    else:
        print(f"Advertencia: DataFrame '{original_file_name}' no encontrado en 'dfs'. Saltando análisis.")
```

Visualizar distribuciones temporales y geográficas.
Nos centraremos en el conjunto de datos «olist_orders_dataset». y podremos representar gráficamente el número de pedidos a lo largo del tiempo.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 'olist_orders_dataset'
orders_df = dfs.get('olist_orders_dataset')

if orders_df is not None:
    print("Conversión de columnas de marca de tiempo de pedido a fecha y hora....")
    date_columns = [
        'order_purchase_timestamp',
        'order_approved_at',
        'order_delivered_carrier_date',
        'order_delivered_customer_date',
        'order_estimated_delivery_date'
    ]
    for col in date_columns:
        orders_df[col] = pd.to_datetime(orders_df[col], errors='coerce')

    # Plot daily order counts
    daily_orders = orders_df['order_purchase_timestamp'].dt.date.value_counts().sort_index()

    plt.figure(figsize=(15, 7))
    daily_orders.plot(kind='line')
    plt.title('Daily Order Counts (Purchase Timestamp)')
    plt.xlabel('Date')
    plt.ylabel('Number of Orders')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Optional: Plot monthly order counts
    monthly_orders = orders_df['order_purchase_timestamp'].dt.to_period('M').value_counts().sort_index()
    plt.figure(figsize=(15, 7))
    monthly_orders.plot(kind='bar')
    plt.title('Monthly Order Counts (Purchase Timestamp)')
    plt.xlabel('Month')
    plt.ylabel('Number of Orders')
    plt.grid(axis='y')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

else:
    print("Error: 'olist_orders_dataset' DataFrame no encontrado.")
```

Análisis de la distribución geográfica utilizando el conjunto de datos `olist_geolocation_dataset`

```python
# 'olist_geolocation_dataset'
geolocation_df = dfs.get('olist_geolocation_dataset')

if geolocation_df is not None:
    print("Visualización de datos geográficos...")

    plt.figure(figsize=(12, 10))
    sns.scatterplot(x='geolocation_lng', y='geolocation_lat', data=geolocation_df, alpha=0.1, s=5)
    plt.title('Distribución geográfica de los códigos postales')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.grid(True)
    plt.show()

    # Top 20 most frequent cities
    plt.figure(figsize=(15, 7))
    geolocation_df['geolocation_city'].value_counts().head(20).plot(kind='bar')
    plt.title('Las 20 ciudades con mayor frecuencia de geolocalización')
    plt.xlabel('City')
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Distribution by state
    plt.figure(figsize=(12, 6))
    geolocation_df['geolocation_state'].value_counts().plot(kind='bar')
    plt.title('Distribución de geolocalización por estado')
    plt.xlabel('State')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()

else:
    print("Error: 'olist_geolocation_dataset' DataFrame no encontrado.")
```
